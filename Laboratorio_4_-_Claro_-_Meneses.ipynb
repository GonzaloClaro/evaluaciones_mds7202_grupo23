El rector, basado en su experiencia, cree fuertemente que el nivel educacional y la etnia de los padres influyen en las notas que obtienen sus hijos.

### 1.3.4 Tabla de Resumen [1 punto]

Para generar la tabla de resumen:

- [ ] Calcular el promedio de las notas y guardarlo en una variable `GPA` (grade point average).
- [ ] Hacer una simplificación a través de un mapeo (investigar el método `map()`) de la variable `parental level of education`  según la siguiente conversión: 

        some high school -> school
        some college -> school
        high school -> school
        bachelor's degree -> college
        associate's degree -> college
        master's degree -> postgraduate

  Los resultados de este mapeo deben ser guardados en la columna `simple parental level of education`.


- [ ] Agregar según 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener el promedio de las notas. 
- [ ] Agregar según 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener un conteo de los alumnos en cada grupo y agregarlos como una nueva fila count.
- [ ] Obtener el porcentaje de alumnos con respecto al total. Los porcentajes deben ser strings que contienen la frecuencia de cada grupo con respecto al total y deben ser terminados en '%'. 


Utilizar la tabla de resultados esperados como guía para desarrollar este punto.

````python
df_grades
````

````python
#se dropea el indice
df_grades_parents = df_grades.drop(["index"], axis = 1)

#calculamos el promedio de los test y se guarda como GPA
df_grades_parents['GPA'] = (df_grades_parents['math score'] + df_grades_parents['reading score'] + df_grades_parents['writing score'])/3

#generamos un diccionario para la conversion y se simplifica mediante la funcion map(), para luego guardarlo en la columna parental level of education
conversion = {'some high school': 'school', 'some college': 'school', 'high school': 'school', "bachelor's degree": 'college', "associate's degree": 'college', "master's degree": 'postgraduate'}
df_grades_parents["simple parental level of education"] = df_grades_parents["parental level of education"].map(conversion)

#agregamos segun race/ethnicity y simple parental level of education para calcular el promedio
df_etnia = df_grades_parents.groupby(['race/ethnicity', 'simple parental level of education']).mean()
#agregamos segun race/ethnicity y simple parental level of education para obtener un conteo de los alumnos en cada grupo y agregarlos como una nueva fila count
df_etnia["conteo"] = df_grades_parents.groupby(['race/ethnicity', 'simple parental level of education']).names.count()
#redondeamos todos estos valores al segundo decimal
df_etnia = df_etnia.round(2)

#obtenemos el porcentaje de alumnos con respecto al total
df_etnia["porcentaje"] = (100*df_etnia["conteo"]/df_etnia["conteo"].sum()).round(2)
df_etnia["porcentaje"] = df_etnia["porcentaje"].astype(str) + " %"

display(df_etnia.sort_values(by=['race/ethnicity','GPA'], ascending=False))
````

````python
other_grades =  pd.read_csv(path+"other_grades.csv").set_index('names')

df_grades1 = df_grades.set_index('names').join(other_grades, how='outer')
display(df_grades1)
````

### Left

````python
df_grades2 = df_grades.set_index('names').join(other_grades, how='left')
display(df_grades2)
````

### Right

````python
df_grades3 = df_grades.set_index('names').join(other_grades, how='right')
display(df_grades3)
````

### Inner

````python
df_grades4 = df_grades.set_index('names').join(other_grades, how='inner')
display(df_grades4)
````

### Justificación

> **Justificación:**


```
Se pueden utilizar 2 métodos, el de left join y el de inner join. Esto ya que ambos cumplen con la cantidad de individuos del dataset original, es decir, no agregan individuos del nuevo dataset, sino que completan las notas de los que ya estaban, por eso tienen 875 filas. Ambos métodos utilizan la intersección se datos, por eso no agregan datos, pero left join rellena las columnas sin datos con NaNs (lo cual no pasa ya que el primer dataset ingresado contiene las 875 filas de la intersección).

Los métodos de right join y outer join son erróneos ya que agregan los alumnos que están en el nuevo dataset, por eso tiene 1000 filas. Al contrario que left join, el método de right join no sirve ya que el dataset ingresado en la izquiera no contiene todos los datos de los alumnos del dataset de la derecha, por ende rellenará con NaNs las columnas de los nuevos alumnos (que no deberían aparecer). Por otra parte, el método de outer join no funciona, ya que utiliza la unión de ambos conjuntos, rellenando con NaNs los valores de los nuevos alumnos.
```
